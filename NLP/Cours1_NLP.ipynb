```python
#Lire le fichier  SMSSpamCollection.tsv par la methode pandas
import pandas as pd
data = pd.read_csv("C:\\Users\\PC\\Desktop\\Cours_INGC3\\NLP\\SMSSpamCollection.tsv", sep ="\t",
                  names = ['label', 'body_text'], header = None)
```

```python
#Importer toutes les ponctuation dans l e bibliotheque string
import string
string.punctuation
```

```python
#Supprimer les ponctuation dans le texte
def remove_punct(text): 
    text_no_punct = ''.join([char  for char in text if char not in string.punctuation])
    return text_no_punct
data['body_text_no_punct'] = data['body_text'].apply(lambda x: remove_punct(x))
data.head()
```

```python
import  re 
def tokenizer(text):
    tokens = re.split('\W+', text)
    return  tokens
data['body_text_tokenized'] = data['body_text_no_punct'].apply(lambda x: tokenizer(x))
data.head()
```

```python
#importer le nltk puis faire telechargement 
import nltk
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    text_no_stop = [ word for word in tokenized_list if word not in stopwords]
    return text_no_stop
data['body_text_nostop'] = data['body_text_tokenized'].apply(lambda x: remove_stopwords(x))
data.head()
```

```python
ps = nltk.PorterStemmer()
def stemming(tokenized_text):
    text = [ps.stem(word) for  word in tokenized_text]
    return text
data['body_te xt_stemmed'] = data['body_text_nostop'].apply(lambda x: stemming(x))
data.head(10)
```

```python
wn = nltk.WordNetLemmatizer()
def lemmatize(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

data[ 'body_text_lemmaized'] = data['body_text_nostop'].apply(lambda x: lemmatize(x))
data.head(10)
```

```python
def clean_text(text):
    text = "".join([char for char in text if char not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if word not in stopwords]
    return text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer(analyzer = clean_text)
x_counts = count_vect.fit_transform(data['body_text'])
print(x_counts.shape)
print(x_counts.toarray())
```
