```python
#Lire le fichier  WordEmbedding _02/src.txt par la methode pandas
import pandas as pd
data = pd.read_csv("WordEmbedding _02/src.txt", sep ="\t",
                  names = ['body_text'], header = None)
data.head()
```

```python
import string 
string.punctuation


def remove_punct(text):
    text_no_punct = ''.join([char for char in text if char not in 
                             string.punctuation])
    return text_no_punct

data['body_text_nopunct'] = data['body_text'].apply(lambda x: remove_punct(x))
data.head(10)                            
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('french')
print(stopwords)
```

```python
len(stopwords)
```

```python


#data.split(' ')
```

```python
import re
def tokenizer(text):
    tokens = re.split('\W+',text)
    return tokens

data['body_text_tokenized']=data['body_text_nopunct'].apply(lambda x: tokenizer(x))
data.head()                            
```

```python
def remove_stopwords(tokenized_list):
    text_no_stop = [word  for word in tokenized_list if word not in stopwords]
    return text_no_stop

data['body_text_nostop'] = data['body_text_tokenized'].apply(lambda x: 
                                                             remove_stopwords(x))
data.head() 
```

## Step 2: Entrianement du model

**a. definissez les hyperparametres du model Word2Vec de gensim.models (embedding_size=300, window_size=3,min_count=2)**

**b. Entrainer le models sur le corpus d'apprentissage**

```python
from gensim.models import Word2Vec
# define training data
sentences = data['body_text_nostop'].tolist()
EPOCHS = 300
model = Word2Vec(sentences, min_count=2, vector_size=300, window = 3, sg=1, epochs=EPOCHS)
```

```python
#Resumé du model
print(model)
```

```python
#vocabulaire
words = list(model.wv.key_to_index)
print(words)
```

```python
#Acceder au vecteur d'un mot
word = 'important'
print(model.wv[word])
```
