# Word Embeddings

L'incorporation de mots "est une famille de techniques de traitement du langage naturel visant à mapper la signification sémantique dans un espace géométrique. Cela se fait en associant un vecteur numérique à chaque mot d'un dictionnaire, de telle sorte que la distance entre deux vecteurs quelconques capturerait une partie de la relation sémantique entre les deux mots associés. L'espace géométrique formé par ces vecteurs est appelé un espace d'intégration.

Par exemple, «noix de coco» et «ours polaire» sont des mots sémantiquement très différents, de sorte qu'un espace d'intégration raisonnable les représenterait comme des vecteurs très éloignés. Mais «cuisine» et «dîner» sont des mots liés, ils doivent donc être imbriqués l'un près de l'autre.


Les embeddings de mots sont calculés en appliquant des techniques de réduction de dimensionnalité à des ensembles de données de statistiques de cooccurrence entre des mots dans un corpus de texte. Cela peut être fait via des réseaux de neurones (la technique "word2vec"), ou via une factorisation matricielle.

## plot vector space

```python
from sklearn.decomposition import PCA
from matplotlib import pyplot 
X = model.wv[new_model.wv.index_to_key]
pca = PCA(n_components=2)
result = pca.fit_transform(X)
# create a scatter plot of the projection
pyplot.figure(figsize=(12,8))
pyplot.scatter(result[:, 0], result[:, 1])
words = list(new_model.wv.key_to_index)
for i, word in enumerate(words):
    pyplot.annotate(word, xy=(result[i, 0], result[i, 1]))
pyplot.show()
```
