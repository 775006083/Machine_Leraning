##### L'implementation de cette classe fut grace aux démonstration mathématiques issues de ce PDF : https://www.robots.ox.ac.uk/~az/lectures/ml/lect2.pdf

```python
class SVMs:
  def __init__(self,num_iters, C, lr): 
    self.num_iters= num_iters
    self.lr= lr
    self.w= None 
    self.C = C # pour la marge souple

  def add_ones(self, x):
    ### Ajouter une colonne de biais c-a-d une colonne de valeur 1
    x_new = np.concatenate((x, np.ones((x.shape[0], 1))), axis=1)
    return x_new
    #### END CODE ####

  def h_function(self, x, w):
    return  x.dot(w) # retourne une matrice Y pour les prédictions de classe de x

  def w_update(self,x, y_true):
    λ = 2/(x.shape[0]*self.C) # λ = 2/NC
    y_hat = self.h_function(x, self.w)
    for i in range(y_true.shape[0]):
      if y_hat[i,0]*y_true[i,0] < 1:
        x_i = x[i,:].reshape((1,x.shape[1]))
        self.w = self.w - self.lr*(λ*self.w - y_true[i,:]*(x_i.T))
      else:
        self.w = self.w - self.lr*λ*self.w
    return True

  #Entrainement du model
  def fit(self, x, y): # prend en parametre x_train et y_train
    x = self.add_ones(x) # Add ones to x for biais
    y[y == 0] = -1 # remplacer les valeurs 0 en -1
    self.w = np.zeros((x.shape[1],1))  # Initialize w to zeros vector 
    ### find best w values ###
    print("Training")
    for i in range(0,self.num_iters) :
       self.w_update(x, y)
    print("...Training Completed")
    
  
  def predict(self, x):
    y_hat = self.predict_proba(x) # Get probability of x
    # Convert proba to 0 or 1
    y_hat = np.where(y_hat >= 0, 1, y_hat)
    y_hat = np.where(y_hat < 0,  0, y_hat)
    return y_hat
  
  def getW(self):
    return self.w.copy()
  
  def predict_proba(self, x):
    x= self.add_ones(x) # Apply add ones to x
    y_pred = self.h_function(x,self.w)  # Predict proba with h_function
    return y_pred
  
  def accuracy(self, y_test, y_pred): # pour le pourcentage de bonne prédiction
    return (np.sum(y_pred == y_test)/len(y_test))*100  
```

### S3VM light

```python
class S3VM:
  def __init__(self, x_un, num_iters, C, C_star, lr): 
    self.x_un = x_un
    self.num_iters =  num_iters
    self.C = C
    self.lr = lr
    self.C_star = C_star
    self.svm_classifier = None
    self.y_un_true = None
   
  def train_svm(self, x_train, y_train):
    svm_classifier = SVMs(num_iters= self.num_iters, C= self.C, lr= self.lr)
    svm_classifier.fit(x_train, y_train.copy())
    # A présent le classifier est pret a etre utiliser avec les valeurs de x_un
    return svm_classifier

  def add_ones(self, x):
    x_new = np.concatenate((x, np.ones((x.shape[0], 1))), axis=1)
    return x_new

  def o_i(self, x_line):
    w = self.svm_classifier.w
    x_i = x_line.reshape((1, x_line.shape[0]))[0]
    return  x_i.dot(w)[0]

  def V(self, y, o_i):  # Hinge loss function
    return max(0,  1 - y*o_i)

  def reverse_cndtion(self, y, x, y_sup, x_sup):
    a = 0; b = 0
    a = self.V(y, self.o_i(x)) + self.V(y_sup, self.o_i(x_sup))
    b = self.V(y, self.o_i(x_sup)) + self.V(y_sup, self.o_i(x))
    rst = True if  a > b else False
    return rst

  def reverse(self, x, y):
    len = y.shape[0]; temp = 0
    x_un = self.add_ones(x)
    for i in range(len):
      if i!= len -1:
        if y[i,0] != y[i+1,0] and self.reverse_cndtion(y[i,0], x_un[i,:], y[i+1,0], x_un[i+1,:]):
          temp = y[i,0]; y[i,0] = y[i+1,0] ; y[i+1,0] = temp
    if y[0,0] != y[len-1,0] and self.reverse_cndtion(y[0,0], x_un[0,:], y[len-1, 0], x_un[len-1,:]) :
      temp = y[0,0]  ; y[0,0] = y[len-1,0] ; y[len-1,0] = temp     
    return y.copy()
 
  def boucle(self, x_un, y_un):
    c_tilde = 1e-5*self.C_star
    while c_tilde < self.C_star:
      y_un = self.reverse(x_un, y_un)
      c_tilde += 1e-5
    return y_un
   
  def get_label(self):
    return self.y_un_true
  
  def fit(self,x_train, y_train):
    self.x_train = x_train
    self.y_train = y_train
    self.svm_classifier = self.train_svm(x_train, y_train)
    y_un = self.svm_classifier.predict(self.x_un) # predict unlabel data with SVM model obtained by  the trainset
    y_un[y_un== 0] = -1 # remplacer 0 par -1
    self.y_un_true = self.boucle(self.x_un.copy(), y_un.copy())
    self.y_un_true[self.y_un_true== -1] = 0
    pass
   
  def accuracy(self, y_true, y_pred): # pour le pourcentage de bonne prédiction
    return (np.sum(y_pred == y_true)/len(y_true))*100 
```
